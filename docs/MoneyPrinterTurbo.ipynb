## 3. Launch Application and Generate Public URL

Now we'll start the Streamlit server and create an ngrok tunnel to make it accessible online:

```python
import glob
import os

# 搜索所有可能的输出位置
paths = [
    '/content/MoneyPrinterTurbo/output',
    '/content/MoneyPrinterTurbo',
    '/content/output',
    '/content',
]

for p in paths:
    if os.path.exists(p):
        print(f"\n=== {p} ===")
        try:
            files = os.listdir(p)
            for f in files[:20]:
                full = os.path.join(p, f)
                size = os.path.getsize(full) if os.path.isfile(full) else "DIR"
                print(f"  {f} ({size})")
        except Exception as e:
            print(f"  读取失败: {e}")

# 也搜索所有mp4文件
print("\n=== 所有mp4文件 ===")
mp4s = glob.glob('/content/**/*.mp4', recursive=True)
for f in mp4s[:20]:
    print(f)
```

```python
import subprocess
import time

print("🚀 Starting MoneyPrinterTurbo...")
# Start Streamlit server on port 8501
streamlit_proc = subprocess.Popen([
    "streamlit", "run", "./webui/Main.py", "--server.port=8501"
])

# Wait for the server to initialize
time.sleep(5)

print("🌐 Creating ngrok tunnel to expose the MoneyPrinterTurbo...")
public_url = ngrok.connect(8501, bind_tls=True)

print("✅ Deployment complete! Access your MoneyPrinterTurbo at:")
print(public_url)
```
